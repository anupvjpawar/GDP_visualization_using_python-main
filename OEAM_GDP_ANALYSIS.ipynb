```python
import os
```

```python
gsva = pd.DataFrame()
```

```python
path = 'C:\\Users\\asus\\Desktop\\case study\\Case Study 3_ GDP Analysis with Niti Ayog\\Datasets\\GSVA\\'
```

```python
listoffiles = os.listdir(path)
```

```python
for filenames in listoffiles:
    for statesname in states:
        if statesname in filenames:
            filepath = path+filenames
            data = pd.read_csv(filepath,encoding='ISO-8859-1')
            data = data[['Item','2014-15']]
            data = data.T
            data.columns = data.iloc[0]
            data = data[1:]
            data['State'] = statesname
            gsva = pd.concat([data,gsva])
                
```

```python
gsva.set_index('State',inplace=True)
```
